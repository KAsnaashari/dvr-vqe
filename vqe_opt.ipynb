```python
import numpy as np
from lib.utils import *
from lib.local_utils import *
from qiskit.algorithms.optimizers import L_BFGS_B, SLSQP
import pylab

import matplotlib.pyplot as plt
from matplotlib import rc
%matplotlib inline
%load_ext autoreload

plt.rcParams["figure.figsize"] = (8,6)
plt.rcParams["font.size"] = 14
plt.rcParams["figure.dpi"] = 300
rc('text', usetex=True)
rc('font',**{'family':'serif','serif':['Times']})
```

```python
excited_states = scratch_dir + 'excited_states/'
```

```python
model_dir = excited_states + 'arhcl_twolocal_opt_1(2)/'
vqe1 = np.load(model_dir + 'vqe.npz', allow_pickle=True)
vals1 = vqe1['vals']
counts1 = vqe1['counts']

model_dir = excited_states + 'arhcl_twolocal_opt_4(2)/'
vqe2 = np.load(model_dir + 'vqe.npz', allow_pickle=True)
vals2 = vqe2['vals']
counts2 = vqe2['counts']

ref = -115.265
ref_32 = -115.17796582
```

```python
for i, optimizer in enumerate(['L-BFGS-B', 'SLSQP']):
    # print(optimizer, vals1[i], abs(ref - vals1[i]))
    pylab.plot(counts1[i], vals1[i], label=optimizer)

print(abs(ref - ref_32))
pylab.axhline(ref_32, ls='--', c='blue', label='Numerical g.s.')
pylab.axhline(ref, ls='dotted', c='black', label='Converged g.s.')
pylab.xlabel('Expectation evaluations')
pylab.ylabel('Energy (cm$^{-1}$)')
# pylab.title('Energy convergence for various optimizers')
# pylab.xlim(800, 1000)
pylab.ylim(-116, -90)
pylab.legend(loc='upper right')
```
